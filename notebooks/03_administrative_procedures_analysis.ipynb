Controlla l'anagrafica IPA e prepara l'analisi dei procedimenti. La tabella dei procedimenti rimane vuota finché non vengono integrati cataloghi ufficiali.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
from IPython.display import display

from scripts.config import CLEAN_DATA_DIR, RAW_DATA_DIR, SOURCE_FILE_NAMES
from scripts.src.clean_administrative_data import clean_administrative_data
from scripts.src.download_administrative_data import download_administrative_data
from scripts.utils import read_dataframe
```

```python
raw_ipa_path = RAW_DATA_DIR / SOURCE_FILE_NAMES["ipa_entities_raw"]
if not raw_ipa_path.exists():
    download_administrative_data()

authorities, procedures = clean_administrative_data()
print(f"Enti IPA: {len(authorities):,}")
print(f"Procedimenti: {len(procedures):,}")
```

```python
authority_summary = (
    authorities
    .groupby(["territorial_level", "authority_type"], dropna=False)
    .size()
    .rename("authority_count")
    .reset_index()
    .sort_values("authority_count", ascending=False)
)
display(authority_summary.head(30))
```

```python
missingness = (
    authorities[["website", "digital_address", "municipality_code"]]
    .isna()
    .mean()
    .rename("missing_share")
    .reset_index(names="field")
)
display(missingness)
```
